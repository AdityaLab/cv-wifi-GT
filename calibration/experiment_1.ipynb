```python

def get_inf_redn(pol_df):
    
    pol_df['total_inf'] = pol_df['external_inf'] + pol_df['internal_inf']

    total_pol = pol_df['total_inf'].iloc[pol_df.shape[0]-1]
    inf_redn = (total_nointervention-total_pol)/total_nointervention
    
    return inf_redn*100
```

```python
max_nointervention = all_df['infected'].max()


def get_peak_redn(pol_df):
    max_pol = pol_df['infected'].max()
    peak_redn = (max_nointervention-max_pol)/max_nointervention
    
    return peak_redn*100
```
